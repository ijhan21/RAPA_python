```python
num = 60
full_len = 2*num -1
for i in range(1,num+1):
    print("{:^{}}".format("#"*(2*i-1),full_len))
for i in range(1,num+1):
    print("{:<{}}".format("#"*(2*i-1),full_len))
for i in range(1,num+1):
    print("{:>{}}".format("#"*(2*i-1),full_len))
```
