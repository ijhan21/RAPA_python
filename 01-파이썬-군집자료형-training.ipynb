# 실습문제 2차원 리스트 만들기

```python
myList = list()
#1. 리스트에 1~12까지의 숫자 넣기
row = 3
col = 4
for i in range(1,row*col+1):
    myList.append(i)
#2. 리스트에서 숫자를 꺼내서 출력하기
for i in range(0, row):
    for j in range(0, col):
        print(f"{myList[i*col+j]:2}",end=" ")
    print()
```

# 아이돌 성적

```python
idols = ['BTS', '트와이스', '블랙핑크', '뉴이스트']
scores = [int(input(f"'{idol}'의 성적 입력:")) for idol in idols]    
print("""-----------------------------------------
아이돌 성적 결과
-----------------------------------------""")

result=["PASS" if scores[i]>=80 else "FAIL" for i in range(len(scores))]

for i, v in enumerate(idols):
    print(f"'{v}'의 성적은 {scores[i]}점: {result[i]}")
```

# 소요 날짜 계산

```python
dates = [31,28,31,30,31,30,31,31,30,31,30,31]
# 날짜 입력
now = input("YYYYMMDD : ")
year = int(now[:4])
mon = int(now[4:6])
day = int(now[6:])

# 윤년 여부 파악
check_year = year%400 ==0 or (year%100 !=0 and year%4 ==0)

# 일수 계산
if check_year:
    dates[1]=29 # 윤년 2월 일수 수정
count_days = sum(dates[:mon-1])+day
print("-"*30)
print(f"{year}/1/1 ~ {year}/{mon}/{day} : {count_days}일")
```

```python
prompt = """
-------------------------
1. 단어 등록
2. 단어 검색
3. 모든 단어 보기
0. 종료하기
-------------------------
"""
word_dict = dict()

while True:
    num = input(prompt)
    if num not in '0123': continue
    choice = int(num)
    if choice == 0:
        print("프로그램 종료")
        break
    elif choice == 1: # 단어 등록
        w = input("새 단어 : ")
        v = input(f"{w}의 뜻 : ")
        word_dict.setdefault(w,v)
        print("새로운 단어가 등록되었습니다.")
    elif choice == 2: # 단어 검색
        w = input("검색할 단어 : ")
#         if w in word_dict.keys():
        if w in word_dict:
            print(f"{w} : {word_dict[w]}")
        else:
            print("=== 미등록 단어입니다 ===")
    elif choice == 3: # 모든 단어 출력
        print("-------------------------")
        print("등록된 단어 목록")
        for k, v in word_dict.items():
            print(f"{k} : {v}")
    else:
        print("번호를 다시 선택해 주세요")
```

# 퀴즈풀기

```python
import random
prompt = """
-------------------------
1. 단어 등록
2. 단어 검색
3. 모든 단어 보기
4. 퀴즈 풀기
0. 종료하기
-------------------------
"""
word_dict = dict()

while True:
    num = input(prompt)
    if num not in '01234': continue
    choice = int(num)
    if choice == 0:
        print("프로그램 종료")
        break
    elif choice == 1: # 단어 등록
        w = input("새 단어 : ")
        v = input(f"{w}의 뜻 : ")
        word_dict.setdefault(w,v)
        print(f"새로운 단어가 {w}: {v} 등록되었습니다.")
    elif choice == 2: # 단어 검색
        w = input("검색할 단어 : ")
#         if w in word_dict.keys():
        if w in word_dict:
            print(f"{w} : {word_dict[w]}")
        else:
            print("=== 미등록 단어입니다 ===")
    elif choice == 3: # 모든 단어 출력
        print("-------------------------")
        print("등록된 단어 목록")
        for k, v in word_dict.items():
            print(f"{k} : {v}")
    elif choice == 4: # 퀴즈 풀기
        question = random.choice(list(word_dict.keys()))        
        while True:
            print("-------------------------")
            ans = input(f"{word_dict[question]}(은)는 영어로?(모르면 0 입력) : ")
            if ans == question:
                print("정답!!!")
                break
            elif ans == '0':
                print("다음에 또 해봅시다!")
                break                    
            else:
                print("Try Again")
        
    else:
        print("번호를 다시 선택해 주세요")
```

```python
class Celsius:
    def __init__(self):
#         self.temperature = 0
        pass


    def to_fahrenheit(self):
        return (self.temperature * 1.8) + 32

    @property
    def temperature(self):
        print("Getting value")
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        if value < -273:
            raise ValueError("Temperature below -273 is not possible")
        print("Setting value")
        self._temperature = value
```
